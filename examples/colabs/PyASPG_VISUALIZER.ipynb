```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
```

```python
DIR_PATH = "/content/drive/MyDrive/PyASPG/1_simulation_results"
```

```python
generator_df = pd.read_csv(DIR_PATH + "/generators.csv")
generator_df.head(100)
```

```python
# Plotting the 'output' feature value over 'timestep'
resampled_data = generator_df.iloc[::20]

# Create figure
fig = go.Figure()

# Add trace for output over time
fig.add_trace(go.Scatter(x=resampled_data['timestep'], y=resampled_data['output'],
                         mode='lines+markers', name='Output', line=dict(color='blue')))

# Update layout
fig.update_layout(
    title='GENERATOR -- Power Output over Time (Every 20 Timesteps)',
    xaxis_title='Timestep (Minutes)',
    yaxis_title='Output (Watts)',
    template='plotly_white'
)

fig.show()
```

```python
prosumers_df = pd.read_csv(DIR_PATH + "/prosumers.csv")
prosumers_df.head(100)
```

```python
# Total Net Power Over Time
total_net_power = prosumers_df.groupby('timestep')['net_power'].sum().reset_index()

fig1 = px.line(total_net_power, x='timestep', y='net_power', title='Total Net Power Over Time')
fig1.show()

# Stored Energy Over Time for Each Prosumers
fig2 = px.line(prosumers_df, x='timestep', y='stored_energy', color='name', title='Stored Energy Over Time for Each Prosumers')
fig2.show()

# Received Power Distribution
fig3 = px.histogram(prosumers_df, x='received_power', title='Distribution of Received Power')
fig3.show()

# Net Power Before and After
fig4 = go.Figure()
fig4.add_trace(go.Scatter(x=prosumers_df['timestep'], y=prosumers_df['net_power_before'], mode='lines+markers', name='Net Power Before'))
fig4.add_trace(go.Scatter(x=prosumers_df['timestep'], y=prosumers_df['net_power'], mode='lines+markers', name='Net Power After'))
fig4.update_layout(title='Net Power Before and After', xaxis_title='Timestep', yaxis_title='Net Power')
fig4.show()
```

```python
# Energy Flow for a Specific Prosumer (e.g., H1)
prosumer = prosumers_df[prosumers_df['name'] == 'H1']
fig5 = go.Figure()
fig5.add_trace(go.Scatter(x=prosumer['timestep'], y=prosumer['received_power'], mode='lines+markers', name='Received Power'))
fig5.add_trace(go.Scatter(x=prosumer['timestep'], y=prosumer['stored_energy'], mode='lines+markers', name='Stored Energy'))
fig5.add_trace(go.Scatter(x=prosumer['timestep'], y=prosumer['net_power'], mode='lines+markers', name='Net Power'))
fig5.update_layout(title='Energy Flow for Prosumer H1', xaxis_title='Timestep', yaxis_title='Power/Energy')
fig5.show()
```
